## Graphs of Oracle Logical IO load tests
This notebook creates the graphs of performance data collected using SLOB, a workload generator for Oracle databases.  
See also: https://kevinclosson.net/slob/

We don't report the exact CPU model in this doc.  
We omit the configuration of networking and I/O, as not relevant for these tests.  
The servers were installed with RHEL 7.9 and Oracle tests used Oracle 19c (v. 19.17).  

**RAC52:**
 - 20 physical cores (2 sockets, 10 physical cores each), 40 logical cores visible on the OS due to hyperthreading
 - CPU max freq: 2.20 GHz
 - CPU from 2016, L1 caches: 32K + 32K, L2 cache 256K, l# cache 25600K
 - RAM: DDR4, 512 GB
 
**RAC54:**
 - 8 physical cores (2 sockets, 4 physical cores each), 16 logical cores visible on the OS due to hyperthreading
 - CPU max freq: 3.80 GHz
 - CPU from 2019, L1 caches: 32K + 32K, L2 cache 1024K, L3 cache 16896K
 - RAM: DDR4, 768 GB
 
**RAC55:**
 - 16 physical cores (2 sockets, 8 physical cores each), 32 logical cores visible on the OS due to hyperthreading
 - CPU max freq: 4.2 GHz
 - CPU from 2019, L1 caches: 32K + 32K, L2 cache 512K, L3 cache 32768K
 - RAM: DDR4, 1 TB

```python
import pandas as pd
```

```python
PATH = "./data_tests_March2023/"
```

```python
data = []
data.append({'key': 'RAC52', 'val': pd.read_csv(PATH + "RAC52_SLOB_Oracle_LIO.csv", sep = ', ', dtype=float)})
data.append({'key': 'RAC54', 'val': pd.read_csv(PATH + "RAC54_SLOB_Oracle_LIO.csv", sep = ', ', dtype=float)})
data.append({'key': 'RAC55', 'val': pd.read_csv(PATH + "RAC55_SLOB_Oracle_LIO.csv", sep = ', ', dtype=float)})
```

```python
data[0].get('val').columns
```

```python
data[0].get('val').head()
```

```python
# Raw Logical IOPS data
```

```python
import matplotlib.pyplot as plt
plt.clf()
plt.style.use('default')

width = 12
height = 5

node_config = {1: {'name': 'RAC52', 'linestyle': 'solid', 'color': 'black', 'alpha': 1.0, 'linewidth': 3}, 
               2: {'name': 'RAC54', 'linestyle': 'solid', 'color': 'red', 'alpha': 1.0, 'linewidth': 3}, 
               3: {'name': 'RAC55', 'linestyle': 'solid', 'color': 'blue', 'alpha': 1.0, 'linewidth': 3} 
              }

# Job duration
fig, ax = plt.subplots(figsize=(width, height))

for i, measurement in enumerate(data):
    df = measurement.get('val')
    instance = i + 1
    df.plot(kind='scatter', x='Num_concurrent_workers', y='Oracle_Logical_IOPS', ax=ax, 
               linestyle=node_config[instance]['linestyle'],
               color=node_config[instance]['color'],
               alpha=node_config[instance]['alpha'],
               linewidth=node_config[instance]['linewidth'],
               label=node_config[instance]['name'])

ax.set_title('Cumulative Oracle Logical IO (reading blocks from the buffer cache, IOPS)')
ax.set_ylim([0, 20])
ax.set_xlabel('Number of concurrent workers')
ax.set_ylabel('Logical IOPS')
# ax.yaxis.set_label_position("right")
#ax.yaxis.tick_right()
#ax.axhspan(0, 30, facecolor='green', alpha=0.15)
#ax.axhspan(30, 50, facecolor='orange', alpha=0.15)
#ax.axhspan(50, 100, facecolor='red', alpha=0.15)
ax.grid()

plt.show()
```

```python
# Logical IOPS per worker
```

```python
import matplotlib.pyplot as plt
plt.clf()
plt.style.use('default')

width = 12
height = 5

node_config = {1: {'name': 'RAC52', 'linestyle': 'solid', 'color': 'black', 'alpha': 1.0, 'linewidth': 3}, 
               2: {'name': 'RAC54', 'linestyle': 'solid', 'color': 'red', 'alpha': 1.0, 'linewidth': 3}, 
               3: {'name': 'RAC55', 'linestyle': 'solid', 'color': 'blue', 'alpha': 1.0, 'linewidth': 3} 
              }

# Job duration
fig, ax = plt.subplots(figsize=(width, height))

for i, measurement in enumerate(data):
    df = measurement.get('val')
    instance = i + 1
    df['Logical_IOPS_per_worker'] = df['Oracle_Logical_IOPS'] / df['Num_concurrent_workers']
    df.plot(kind='scatter', x='Num_concurrent_workers', y='Logical_IOPS_per_worker', ax=ax, 
               linestyle=node_config[instance]['linestyle'],
               color=node_config[instance]['color'],
               alpha=node_config[instance]['alpha'],
               linewidth=node_config[instance]['linewidth'],
               label=node_config[instance]['name'])

ax.set_title('Oracle Logical IO per worker (shows how IOPS throughput per worker declines with increasing load)')
#ax.set_ylim([0, 20])
ax.set_xlabel('Number of concurrent workers')
ax.set_ylabel('Logical IOPS')
# ax.yaxis.set_label_position("right")
#ax.yaxis.tick_right()
#ax.axhspan(0, 30, facecolor='green', alpha=0.15)
#ax.axhspan(30, 50, facecolor='orange', alpha=0.15)
#ax.axhspan(50, 100, facecolor='red', alpha=0.15)
ax.grid()

plt.show()
```

```python
# Speedup
```

```python
import matplotlib.pyplot as plt
plt.clf()
plt.style.use('default')

width = 12
height = 5

node_config = {1: {'name': 'RAC52', 'linestyle': 'solid', 'color': 'black', 'alpha': 1.0, 'linewidth': 3}, 
               2: {'name': 'RAC54', 'linestyle': 'solid', 'color': 'red', 'alpha': 1.0, 'linewidth': 3}, 
               3: {'name': 'RAC55', 'linestyle': 'solid', 'color': 'blue', 'alpha': 1.0, 'linewidth': 3} 
              }

# Job duration
fig, ax = plt.subplots(figsize=(width, height))

for i, measurement in enumerate(data):
    df = measurement.get('val')
    instance = i + 1
    base = df['Oracle_Logical_IOPS'][0]
    df['speedup'] = df['Oracle_Logical_IOPS'] / base
    df.plot(kind='scatter', x='Num_concurrent_workers', y='Oracle_Logical_IOPS', ax=ax, 
               linestyle=node_config[instance]['linestyle'],
               color=node_config[instance]['color'],
               alpha=node_config[instance]['alpha'],
               linewidth=node_config[instance]['linewidth'],
               label=node_config[instance]['name'])

ax.set_title('Speedup (shows how scalable the workload is: linear scalability at low load then saturation at high load)')
ax.set_ylim([0, 20])
ax.set_xlabel('Number of concurrent workers')
ax.set_ylabel('Speedup')
# ax.yaxis.set_label_position("right")
#ax.yaxis.tick_right()
#ax.axhspan(0, 30, facecolor='green', alpha=0.15)
#ax.axhspan(30, 50, facecolor='orange', alpha=0.15)
#ax.axhspan(50, 100, facecolor='red', alpha=0.15)
ax.grid()

plt.show()
```
